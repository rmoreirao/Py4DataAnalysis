```python
import pandas as pd
```

```python
unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
```

```python
users = pd.read_table('..\\Resources\\pydata-book-master\\ch02\\movielens\\users.dat', sep='::', header=None, names=unames)
```

```python
rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
```

```python
ratings = pd.read_table('..\\Resources\\pydata-book-master\\ch02\\movielens\\ratings.dat', sep='::', header=None, names=rnames)
```

```python
mnames = ['movie_id', 'title', 'genres']
```

```python
movies = pd.read_table('..\\Resources\\pydata-book-master\\ch02\\movielens\\movies.dat', sep='::', header=None, names=mnames)
```

```python
 data = pd.merge(pd.merge(ratings, users), movies)
```

```python
data[:5]
```

```python
 mean_ratings = data.pivot_table('rating', index='title', columns='gender', aggfunc='mean')
```

```python
mean_ratings[:5]
```

```python
ratings_by_title = data.groupby('title').size()
```

```python
ratings_by_title[:10]
```

```python
active_titles = ratings_by_title.index[ratings_by_title >= 250]
```

```python
active_titles
```

```python
mean_ratings = mean_ratings.ix[active_titles]
```

```python
mean_ratings
```

```python
top_female_ratings = mean_ratings.sort_values(by='F', ascending=False)
```

```python
 top_female_ratings[:10]
```

```python
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']
```

```python
sorted_by_diff = mean_ratings.sort_index(by='diff')
```

```python
sorted_by_diff[:15]
```
